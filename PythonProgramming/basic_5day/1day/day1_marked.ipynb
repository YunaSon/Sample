## unpacking / packing

- Packing: 파이썬에서는 기본적으로 TUPLE형태로 패킹 된다. 

• 패킹(packing): 한 변수에 여러 개의 데이터를 할당하는 것.

• 언패킹(unpacking): 한 변수의 데이터를 각각의 변수로 반환하는 것.
